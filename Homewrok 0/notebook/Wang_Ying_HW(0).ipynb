### Matplotlib

(a) Create two one dimensional arrays x and y and plot y vs x, add title, xlabel, ylabel, grid.

```python
x = np.linspace(-5, 5, num=20)
y = np.array([j ** 2 for j in x])
```

```python
plt.plot(x,y)
plt.title("y vs x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

(b) Create multiple arrays and plot them with different styles, add legends, add text/mathematical equations on the plot.

```python
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.legend(["y1","y2","y3"])
plt.text(4,100,"y3=x^3")
plt.text(4,16,"y2=x^2")
plt.text(4,-10,"y1=x")
plt.show()
```

(c) Create multiple subplots, play around with the figure size, figure title, and its font style and font size 

```python
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])
plt.subplot(1,3,1)
plt.plot(x,y1)
plt.text(0,0,"y1=x")
plt.xlabel("x",fontsize=10)   
plt.subplot(1,3,2)
plt.plot(x,y2)
plt.text(0,16,"y2=x^2")
plt.xlabel("x",fontsize=10)
plt.subplot(1,3,3)
plt.plot(x,y3)
plt.text(4,100,"y3=x^3")
plt.xlabel("x",fontsize=10)
plt.figure(figsize=(100,400))
plt.show()
```

(e) Change the limits on x and y axes, **use logarithmic axes to plot**.
